```python
import numpy as np 
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
```

```python
##interpolating data with linear interpolation
```

```python
import matplotlib.pyplot  as plt
```

```python
plt.scatter(x , y)
```

```python
x_new = np.linspace(1,5,10) #new x values creation
y_interp = np.interp(x_new , x, y) ##interpolate y values
print(y_interp)
```

```python
plt.scatter(x_new , y_interp)
```
